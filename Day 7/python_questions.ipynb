Q3. Count digits and letters

```python
from string import ascii_letters, digits


index = 0
st = input()

num_digts = 0
num_letters = 0


while index < len(st):

    ch = st[index]
    if not digits.find(ch) == -1: num_digts += 1
    elif not ascii_letters.find(ch) == -1: num_letters += 1

    index += 1

print("Number of digits: ", num_digts)
print("Number of letters: ",num_letters)
```

Q4. Find first n prime numbers

```python
def is_prime(num: int) -> bool:
    if num == 2:
        return True
    elif num < 2 or num % 2 == 0: return False
    
    for i in range(3, num//2 + 1):
        if num % i == 0:
            return False
    return True

n = int(input())

num = 1
prime_num = 0

while prime_num <= n:
    if(is_prime(num)):
        print(num)
        prime_num += 1
    num += 1
```

Q5. ATM

```python
balance = 0
pin = "3333"
def withdraw(amount):
    global balance
    if amount > balance: return -1
    else:
        balance = balance - amount
        return balance

def deposit(amount):
    global balance
    balance += amount
    return balance


pin_attempts = 0
did_enter_correct_pin = False
while pin_attempts < 3:
    pin_input = input()
    if pin_input == pin:
        did_enter_correct_pin = True
        break
    else:
        print("Entered incorrect pin. Re-enter correct pin")
        pin_attempts += 1

if not did_enter_correct_pin:
    print("Used 3 incorrect attempts. Exiting program")

else: 
    option = 999

    print("Welcome to the system")
    while not option == '4':
        option = input("Choose option:\n1. Display Balance\n2. Withdraw\n3. Deposit\n4. Exit")
        if option == '1':
            print("Balance: ", balance)
        elif option == '2':
            am = int(input("Enter amount to withdraw"))
            new_bal = withdraw(am)
            if new_bal == -1:
                print("Balance exceeds requests withdraw amount")
            else:
                print("Success! New balance: ", new_bal)
        
        elif option == '3':
            am = int(input('Enter amount to deposit'))
            new_bal = deposit(am)
            print("Success! New balance: ", new_bal)
        else:
            print("Exiting system")
```

Q6 Pyramid pattern

```python
def print_pyramid(n):
    for i in range(n):
        for _ in range(n - i - 1):
            print(" ", end="")
        for _ in range(2 * i + 1):
            print("*", end="")
        print()

print_pyramid(5)
```

Q7 Number pattern

```python
def print_number(n):
    for i in range(n):
        for j in range(n - i - 1):
            print(" ", end="")
        for k in range(2 * i + 1):
            print(k + 1, end="")
        print()

print_number(5)
```

Q9. Student mark range validation

```python
marks = int(input("Enter marks"))

if marks > 80 and marks <= 100:
    grade = "A"
elif marks <= 80 and marks >70:
    grade = "B"
elif marks <=70 and marks >60:
    grade = "C"
elif marks <= 60 and marks >= 40:
    grade = "D"
elif marks >= 0 and marks < 40:
    grade = "F"
else:
    grade = "Incorrect marks value. Should be between 0 - 100"

print("Grade: " + grade)
```

Q10. Find roots of a 2nd degree polynomial expression

```python
import math
from numpy import roots

c = int(input("c"))
b = int(input("b"))
a = int(input("a"))

roots_ = roots([c,b,a])
d = math.pow(b,2) - (4 * a * c)
if (d > 0): print("Real and distinct roots")
elif (d == 0): print("Real and equal roots")
else: print("Imaginary or complex roots")

print("Roots: " , roots_)
```
