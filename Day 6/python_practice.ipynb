Q1. WAP to get area of shape

```python
import math


def area(shape, attr):
    if shape == "circle":
        return math.pi * 2 * attr['r'] ** 2
    elif shape == "square":
        return attr['l'] * attr['l']
    elif shape == "rectangle":
        return attr['l'] * attr['b']
    elif shape == "triangle":
        return attr['b'] * attr['h'] * 1/2
    else:
        raise ValueError("Wrong value of shape inputted")

shp = input("Provide name of shape")
if shp == "circle":
    attr = {"r": int(input("Enter radius"))}
print(area(shp, attr))
```

Q2. WAP to check if string is palindrome or not

```python
st = input("Enter string to be checked")
print("Inputted string: " , st)
if st[::-1] == st:
    print("Palindrome")
else:
    print("Not a palindrome")
```

Q3. Replace old string with new string and count the number of characters in old string

```python
from curses.ascii import isalnum


old_str = input("Enter old string")
new_str = input("Enter new string")

old_str = new_str 

ch_num = 0
for token in old_str:
    if isalnum(token):
        ch_num += 1

print("Number of characters: " ,ch_num)
```

Q4. Visualize the following using matplotlib:
